```python
populacao_estimada = pd.read_excel("/content/oficina_nlp/data/populacao_estimada_ibge.xlsx")
populacao_estimada.dropna(inplace=True)
populacao_estimada
```

```python
populacao_estimada = populacao_estimada[uf == "31"]
populacao_estimada
```

Escreva, no bloco de código a seguir, uma função que padroniza um dado texto aplicando duas operações:

1.   Converte todos os caracteres em maiúsculo
2.   Converte todos os caracteres para a codificação ASCII



```python
!pip install unidecode
```

```python
from unidecode import unidecode
```

```python
def padroniza_texto(texto):
  texto = texto.upper()
  texto = unidecode(texto)
  return texto 
  
```

Aplique a função criada na coluna `municipio` da tabela `populacao_estimada`:

```python
populacao_estimada["municipio_padr"] = populacao_estimada["municipio"].apply(padroniza_texto)
populacao_estimada
```

Faça o cruzamento ("merge" ou "join") das tabelas:

```python
populacao_estimada.merge(transferencias_municipais, left_on="municipio_padr", right_on="nome")
```

Observem que a tabela cruzada contém apenas 844 registros. Alguns municípios não foram cruzados.
