Neste exercício vamos detectar CPF em relatórios de acórdãos do TCU e suprimir os dígitos intermediários (substitui-los por asteriscos), deixando somente os três primeiros e os dois últimos dígitos de cada CPF encontrado.

Observação: os números de CPF nos textos a seguir foram gerados aleatoriamente.

```python
acordaos = pd.read_csv("/content/oficina_nlp/data/acordaos_cpf.csv")
acordaos
```

Para o exercício, é suficiente detectar e substituir os números de CPF que constam somente no texto a seguir. A aplicação no restante da base é somente uma replicação.

```python
txt = acordaos.iloc[0]["acordao"]
txt
```

Sugestão: criar a expressão regular de detecção usando o https://regex101.com/.

```python
import re
```

```python
# Complete o código substituindo as sublinhas ("___" e "_")
def mascara_cpf(texto):
  output = re.sub(pattern="CPF [___]{_}.([___]{_}).([___]{_})-[___]{_}", repl=r"CPF XXX.\1.\2-XX", string=texto)
  return(output)
```

```python
mascara_cpf(txt)
```
